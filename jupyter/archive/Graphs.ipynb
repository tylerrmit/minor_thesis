```python
import os
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(os.path.join(os.pardir, 'V1_training.csv'))
df.head(3)
```

```python
df2 = pd.read_csv(os.path.join(os.pardir, 'V2_training.csv'))
df2.head(3)
```

```python
os.chdir(os.pardir)
```

```python
df.set_index('Steps', inplace=True)
df.head(3)
```

```python
df.groupby("Model")["test_mAP"].plot(
    legend=True,
    xlabel="Training Steps",
    ylabel="Evaluation Mean Average Precision (mAP)",
    title = "Detection Box Mean Average Precision vs. Test Dataset during training\n\nGSV Models"
)
plt.savefig("fig_gsv_map.png")
plt.show()
```

```python
df.groupby("Model")["train_TotalLoss"].plot(
    legend=True,
    xlabel="Training Steps",
    ylabel="Training Total Loss",
    title="Total Loss vs. Training Dataset during training\n\nGSV Models"
)
plt.savefig("fig_gsv_loss.png")
plt.show()
```

```python
df2.set_index('Steps', inplace=True)
df2.head(3)
```

```python
df2.groupby("Model")["test_mAP"].plot(
    legend=True,
    xlabel="Training Steps",
    ylabel="Evaluation Mean Average Precision (mAP)",
    title = "Detection Box Mean Average Precision vs. Test Dataset during training\n\nDashcam Models"
)
plt.savefig("fig_dash_map.png")
plt.show()
```

```python
df2.groupby("Model")["train_TotalLoss"].plot(
    legend=True,
    xlabel="Training Steps",
    ylabel="Training Total Loss",
    title="Total Loss vs. Training Dataset during training\n\nDashcam Models"
)
plt.savefig("fig_dash_loss.png")
plt.show()
```
